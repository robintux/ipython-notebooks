```python
from scipy import stats 
```

```python
# http://code.activestate.com/recipes/578231-probably-the-fastest-memoization-decorator-in-the-/
def memoize(f):
    """ Memoization decorator for functions taking one or more arguments. """
    class memodict(dict):
        def __init__(self, f):
            self.f = f
        def __call__(self, *args):
            return self[args]
        def __missing__(self, key):
            ret = self[key] = self.f(*key)
            return ret
    return memodict(f)
```

A [two-variate normal distribution](http://en.wikipedia.org/wiki/Multivariate_normal_distribution):

```python
_two_norm = lambda x,y: exp(-0.5 * (x**2 + y**2)) / (2*pi)
%timeit _two_norm(0,0)
two_norm = memoize(_two_norm)
%timeit two_norm(0,0)
```

```python
x = arange(-1, 1, 0.01)
y = arange(-1, 1, 0.01)
def sample(dist):
    z = array([[dist(i,j) for i in x] for j in y])
    return z
```

```python
def plot_2D(z):
    imshow(z,extent=(min(x), max(x), min(y), max(y)))
    colorbar()
plot_2D(sample(two_norm))
```

Mutation:

$$
f_{t+1}(x,y) = \int_{-1}^{1}{\int_{-1}^{1}{f_{t}(a,b) g(a,b,x,y) da} db}
$$

where $g(a,b,x,y)$ is the probability for transition from $(a,b)$ to $(x,y)$.

A simple choice of $g$ is an [chi-squared distribution](http://en.wikipedia.org/wiki/Exponential_distribution) on the square of the distance between the points:

$
g(a,b,x,y) = P(\chi^2(2) = (x-a)^2 + (y-b)^2)
$

```python
mutation_rv = stats.chi(2)
dist = memoize(lambda a,b,x,y: sqrt((x-a)**2 + (y-b)**2))
g = memoize(lambda a,b,x,y: mutation_rv.pdf( 10*dist(a,b,x,y) ) )
```

```python
plot_2D(sample(lambda x,y: g(0,0,x,y)))
```

```python
f_0 = sample(two_norm)
plot_2D(f_0)
```

$$
f_{t+1}(x,y) = \sum_{a=-1}^{1}{\sum_{b=-1}^{1}{ f_{t}(a,b) g(a,b,x,y) }}
$$

```python
def mutation(f):
    return sample(lambda x1,y1: array([f[i,j] * g(x[i],y[j],x1,y1) for i in range(len(x)) for j in range(len(y))]).sum())
```

```python
f_1 = mutation(f_0)
```

```python
plot_2D(f_1)
```
