```python
ats={}
fp.seek(0)
entry = ''
key = ''
for line in fp:
    if line.startswith('@'):
        # begin
        entry = line
        key = line[line.find('{')+1:line.find(',')]
    elif line=='}\n':
        entry += line
        ats[key] = entry
        key = ''
        line = ''
    else:
        entry += line
fp.close()
```
