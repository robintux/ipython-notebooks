```python
def sequential_search(key,lst):
    for k,v in lst:
        if k==key:
            return v
    return None
```

```python
def max_index(lst, stop_at=-1):
    if stop_at == -1:
        stop_at = len(lst)-1
    m = stop_at
    for i in range(stop_at):
        if lst[i] > lst[m]:
            m = i
    return m
max_index(numbers)
```

```python
def bucket_sort(lst): 
    '''for numbers between 0 and 99, with 10 buckets'''
    buckets = [ [] for _ in range(10) ]
    for x in lst:
        buckets[x/10].append(x)
    sorted_list = []
    for b in buckets:
        if len(b) > 0:
            b = selection_sort(b)
            print b
        sorted_list.extend(b)
    return sorted_list
```

```python
def rec_selection_sort(lst):
    '''recursive selection sort'''
    print lst
    if (len(lst)==1):
        return lst
    m = min_index(lst) # m is index of minimal value
    lst[m], lst[0] = lst[0], lst[m] # v is the minimal value
    return [ lst[0] ] + rec_selection_sort(lst[1:])
```
