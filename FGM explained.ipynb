# Fisher's geometrical model

## Two dimensions

Let's start with a simple case. There are two traits (dimensions) to an individual, $x$ and $y$. The optimal trait combination is $(0,0)$ and the distance of the current trait combination is $d$ - for example, the trait combination can be $(d,0)$.

A mutation occurs, changing the trait combination $(x,y)$ to some $(x',y')$, such that the distance between the previous and the new combinations is $r$. If the direction of the mutation is **random**, but its size ($r$) is positive and non-random, then what are the chances that the new trait combination is better then the previous? 

To demonstrate this, look at the following diagram. $A=(d,0)$ is the current trait combination. The blue circle marks all the combintations that are $r$ away from $A$. The red circle marks all the combinations that are exactly as good as $A$ - they are all $d$ away from the optimum (at $O$). A beneficial mutation is such that occurs on the part of the blue circle that is inside the red circle. The probability that a mutation is beneficial $p$ is the ratio between the blue circle that lays inside the red circle.

Next I will draw a the diagram - the specific choice of $d$ and $r$ are just for the sake of making a good diagram.

```python
d=1.
r=.8
x_high = d-r**2/(2*d)
x_low = d-r
y_low = -sqrt(r**2*(1-r**2/(4*d**2)))
y_high = -y_low
```

```python
xlim((-d-r*.5,d+r*1.5))
ylim((-d-r,d+r))
xlabel('x')
ylabel('y')
axhline(y=0,ls='--',color='gray')
axvline(x=0,ls='--',color='gray')

gca().add_artist(Circle((0,0), d, color='r', fill=False))
gca().add_artist(Circle((d,0), r, color='b', fill=False))
gca().add_artist(Polygon([[d,0],[x_high,y_low],[x_high,y_high]], color="green", fill=False))
gca().add_artist(Polygon([[0,-d-r*.5],[d,-d-r*.5]], color="gray", fill=False))
gca().add_artist(Polygon([[d-r,-d-r*.25],[d,-d-r*.25]], color="gray", fill=False))
scatter([0,d,x_high,x_high],[0,0,0,y_high])

text(-.2,.05,'O')
text(d+.05,0.05,'A')
text(x_high+.025,y_high+.1,'B')
text(x_high,y_low-.15,'D')
text(x_high-.15,.05,'C')
text(d-.15,.025,r'$\theta$')
text(d/2, -d-r*.5+.05, r'$r$')
text(d/2, -d-r*.5-.15, r'$d$')

text(d+r*0.5,d+.25,r'$AC=\frac{r^2}{2d}$')
text(d+r*0.5,d,r'$AB=r$')
text(d+r*0.5,-d-.25,r'$\cos{\theta}=\frac{AC}{AB}$')
text(d+r*0.5,-d-.5,r'$p=\frac{\theta}{\pi}$');
```

#### Explanations:

1. $AC$ is actually $d-x$, where $x$ is the x coordinate of the intersection of the circles:
$$
(x-d)^2 + y^2 = r^2 \\\\
x^2 + y^2 = d^2 \Rightarrow \\\\
x = d-\frac{r^2}{2d}
$$
and therefore $AC=\frac{r^2}{2d}$.
2. $AB$ is the radius of the blue circle.
3. $\theta$ is the angle between $AC$ and $AB$.
5. $p$ is the ratio between the fraction of the blue circle that lies inside the red circle - $2\theta$ and the whole circle - $2\pi$.

The final formula for $p$ is:
$$
p = \frac{\arccos{\frac{r}{2d}}}{\pi}
$$
Because the taylor linear approximation of $arccos$ is:
$$
arccos(x) = \frac{\pi}{2} - x +O(x^3)
$$
If we plug that in $p$, we get:
$$
p = \frac{1}{2}(1-\frac{r}{d\pi})
$$
This is similar to the the formula given by Fisher:
$$ 
p= \frac{1}{2}(1-\frac{r}{d})
$$.

```python
fisher_p = lambda r,d: (1-r/d)/2.0
ram_2d_p = lambda r,d: (acos(r/(2.0*d)))/(pi)
taylor_2d_p = lambda r,d: (1-r/(d*pi))/2.0
```

#### Comparison of results

Here, I set $d=1$ and explore $0<r<d$ to see how good are the approximations:

```python
rs = linspace(0,d,100)
ps = [(fisher_p(r,d), ram_2d_p(r,d), taylor_2d_p(r,d)) for r in rs]
f1,f2,f3 = plot(rs, ps)
f3.set_linestyle('--')
legend(['Fisher','Ram', 'Taylor'], loc=3)
xlabel('mutational radius')
ylabel('probability of beneficial');
```

It's clear that Fisher's formula is far away from the formula we got by analizing the circle diagram, but that the Taylor approximation is very good. However, Fisher discussed a sphere in three dimensions, not a circle in two.

## Three dimensions

When going from the two to the three dimensional we have two spheres, a red one centered at $(0,0,0)$ with radius $d$ and blue one centered at $(d,0,0)$ with radius $x$. Their intersection defines a plane which cuts the blue sphere to create a [spherical cap](http://en.wikipedia.org/wiki/Spherical_cap). The area of this spherical cap has a very simple formula - $2\pi r (r-a)$ where $a$ is the distance from the center of the blue sphere to the intersection plane. This distance $a$ is $d-x$ from above and can be found much in the same way - it is euqal to $\frac{r^2}{2d}$.

Because the entire area of the surface of the blue sphere is $4\pi 4$, the ratio between the area of the spherical cap and the whole sphere, and therefore the probability that a mutation with length $r$ is beneficial, is:
$$
p = \frac{2\pi r (r-\frac{r^2}{2d})}{4\pi r^2} = \frac{1}{2}(1-\frac{r}{2d})
$$
which is again slightly different from Fisher's $\frac{1}{2}(1-\frac{r}{d})$ but less than before.


```python
fisher_p = lambda r,d: (1-r/d)/2.0
ram_3d_p = lambda r,d: (1-r/(2*d))/2.0
```

```python
rs = linspace(0,d,100)
ps = [(fisher_p(r,d), ram_3d_p(r,d)) for r in rs]
f1,f2 = plot(rs, ps)
f3.set_linestyle('--')
legend(['Fisher','Ram'], loc=3)
xlabel('mutational radius')
ylabel('probability of beneficial');
```

## N dimensions

In the case of N dimensions we are looking at two N-1 hyperspheres, but most of the other parameters remain the same.

Following [(Li 2011)](http://scialert.net/abstract/?doi=ajms.2011.66.70), but slightly changed according to [Wikipedia](http://en.wikipedia.org/wiki/Spherical_cap#Hyperspherical_cap), the area of a hyperspherical cap is:
$$
\frac{1}{2} A_n (r) r^{n-1} I_{(2rh-h^2)/r^2}(\frac{n-1}{2},\frac{1}{2})
$$
Again, $h$ is equal to $r-x$ where $x$ is the distance of the intersecting hyperplane from the origin, but there are two new functions here:
- $A_n$ is the area of a hypersphere with radius 1
- $I$ is the [regularized incomplete beta function](http://en.wikipedia.org/wiki/Regularized_incomplete_beta_function#Incomplete_beta_function)


Because we want the ratio of the area of the cap to the area of the entire hypersphere, we divide by $A_n r^{n-1}$ to get
$$
\frac{1}{2} I_{(2rh-h^2)/r^2}(\frac{n-1}{2},\frac{1}{2})
$$.

Let's find $h$:

$$
h = r-d+x \\\\
x = d-\frac{r^2}{2d} \Rightarrow \\\\
h= r(1-\frac{r}{2d})
$$
We use $h$ to find the expression in the subscript of $I$:
$$
\frac{(2rh-h^2)}{r^2} = 1-\frac{r^2}{4d^2}
$$
So we get that the probability of beneficial mutation is:
$$
p = \frac{1}{2} I_{1-r^2/4d^2}(\frac{n-1}{2},\frac{1}{2})
$$
or in full form:
$$
p = \frac{1}{2} \frac{\int\limits_0^{1-r^2/4d^2}{t^{(n-1)/2}\sqrt{1-t} \, d t}}{\int\limits_0^{1}{t^{(n-1)/2}\sqrt{1-t} \, d t}}
$$

Here is a comparison of the previous derived formulas with the general formula:

```python
from scipy.special import betainc
full_p = lambda r,d,n: 0.5 * betainc((n - 1)/2.0, 0.5, 1 - r**2/(4 * d*2))
```

```python
rs = linspace(0,d,100)
ps = [(ram_2d_p(r,d), full_p(r,d,2)) for r in rs]
f1,f2 = plot(rs, ps)
f2.set_linestyle('--')
legend(['Ram','Full'], loc=3)
xlabel('mutational radius')
ylabel('probability of beneficial')
title("2D");
```

```python
rs = linspace(0,d,100)
ps = [(ram_3d_p(r,d), full_p(r,d,3)) for r in rs]
f1,f2 = plot(rs, ps)
f2.set_linestyle('--')
legend(['Ram','Full'], loc=3)
xlabel('mutational radius')
ylabel('probability of beneficial')
title("3D");
```

You can see that this new formula agrees perfectly with the derived formulas from above.

How does the probability change with the number of dimensions?

```python
ns = range(2,1000)
rs = [d/5, d/10,d/100]
ps = [[full_p(r,d,n) for r in rs] for n in ns]
plot(ns, ps)
legend(rs, title="r/d", loc=5)
xlabel('dimensions')
ylabel('probability of beneficial');
```

#### In probability theory language

Because of the [relation](https://en.wikipedia.org/wiki/Beta_function#Incomplete_beta_function) between the incomplete beta function $I_x$ and the cumulative distribution function of the binomial distribution ($I_{1-x}(N-K,K+1)=Pr(X \le K;N,x)$, we can approximate the probability of beneficial mutation using a the binomial CDF. This is only an approximation because the binomial distribution is supported on discrete integers but the incomplete beta functions has a parameter 1/2 which cannot be trastlated to the binomial representation.

```python
from scipy.stats import binom
def binomial_p(r,d,n):
    K = int((n - 3.) / 2.)
    N = n/2. - 1
    P = 1 - r**2/(4. * d**2.)
    return 0.5 * (1 - binom(N,P).cdf(K))
rs = linspace(0,d,100)
ps = [(binomial_p(r,d,100), full_p(r,d,100)) for r in rs]
plot(rs, ps);
legend(['Binom','Beta'], loc=1)
xlabel('Mutational Radius')
ylabel('Probability of beneficial');
```

## Numerical simulation

```python
from scipy.linalg import norm
from random import gauss
```

```python
def random_vector(n,r):
    vec = [gauss(0, 1) for i in range(n)]
    mag = norm(vec)
    return array([r*x/mag for x in vec])
```

```python
def is_beneficial(n, d, r):
    z = zeros(n)
    z[0] = d
    v=random_vector(n,r)
    return norm(z+v) < d
```

```python
def beneficial_probability(n, d, r, replicates=1000):
    return mean([is_beneficial(n,d,r) for _ in range(replicates)])
```

```python
d = 1.0
N = 10000
rs = logspace(-6,0)
plot(rs, [beneficial_probability(2,d,r,N) for r in rs], label="2")
plot(rs, [full_p(r,d,2) for r in rs],'--')
plot(rs, [beneficial_probability(3,d,r,N) for r in rs], label="3")
plot(rs, [full_p(r,d,3) for r in rs],'--')
plot(rs, [beneficial_probability(100,d,r,N) for r in rs], label="100")
plot(rs, [full_p(r,d,100) for r in rs],'--')
legend(title="Dimensions", loc=4)
xlabel('Mutational Radius')
ylabel('Probability of beneficial');
```

```python
N = 10000
d = 1.0
ns = range(2,1000,10)
rs = [d/5, d/10,d/100]
ps = [[beneficial_probability(n,d,r,N) for r in rs] for n in ns]
plot(ns, ps)
ps = [[full_p(r,d,n) for r in rs] for n in ns]
plot(ns, ps,'--')
legend(rs, title="r/d", loc=5)
xlabel('dimensions')
ylabel('probability of beneficial');
```
