```python
def simulate(pop_size, mean, low_var, high_var, threshold, generations):
    p = [(pop_size / 2, pop_size / 2)]
    for g in range(generations):
        plow = normal(mean, low_var, p[-1][0])
        phigh = normal(mean, high_var, p[-1][1])
        flow, fhigh = (plow > threshold).sum(), (phigh > threshold).sum()
        ftotal = float(flow + fhigh)
        f = (flow / ftotal, fhigh / ftotal)
        p.append(multinomial(sum(p[-1]), f))
        if p[-1][0] == 0 or p[-1][1] == 0:
            break
    p = array(p)
    return p
```

```python
p = simulate(1e6, 90, sqrt(30), sqrt(45), 100, 10)
```
