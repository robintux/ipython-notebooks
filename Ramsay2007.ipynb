## The FitzHugh-Nagumo neural spike potential equations

$$
\frac{dV}{dt} = c \Big( V - \frac{V^3}{3}+R \Big) + u(t) \\
\frac{dR}{dt} = -\frac{1}{c} (V -a +bR)
$$

```python
%pylab inline
import odespy
import seaborn as sns
sns.set_style('white')
```

```python
def dydt(y,t,a,b,c,u):
    V,R = y
    dVdt = c*(V-V**3/3+R)+u(t)
    dRdt = -(V-a+b*R)/c
    return dVdt, dRdt
```
